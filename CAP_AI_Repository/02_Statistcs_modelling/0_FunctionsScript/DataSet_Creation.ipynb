# CREATE EWS DATASET

```python
d_16_18_mrt = pickle.load(open(path + 'df_ts_2016_18.pickle','rb'))
d_19_20_mrt = pickle.load(open(path + 'df_ts_2019_20.pickle','rb'))
```

```python
def create_ews_1d_dataset(df):
    df_ews = df[df['no_sample_series'] == 143]
    target_1d_ews =list(dict(zip(df_ews['admission_id'], df_ews['ews'])).values())
    df_ = df[df['no_sample_series'] < 48].drop(columns = ['ews', 'no_sample_series'])
    df_ = df.astype(dict([(feat, 'float') for feat in df.columns[1:]]))
    return [df_, target_1d_ews]
```

```python
ews_16_18_data = create_ews_1d_dataset(d_16_18_mrt )
ews_19_20_data = create_ews_1d_dataset(d_19_20_mrt )

pickle.dump(ews_16_18_data, open(path + 'df_ts_2016_18_ews_1d.pickle', 'wb'))
pickle.dump(ews_19_20_data, open(path + 'df_ts_2019_20_ews_1d.pickle', 'wb'))
```

# CREATE LOS DATASET

```python
d_16_18_mrt = pickle.load(open(path + 'df_ts_2016_18.pickle','rb'))
d_19_20_mrt = pickle.load(open(path + 'df_ts_2019_20.pickle','rb'))
```

```python
def create_los_1d_dataset(df, df_admin):
    admins = df['admission_id'].unique().tolist()
    dict_los = dict(zip(df_admin['admission_id'], df_admin['lengthofstay']))
    target_1d_los = []
    for adm in admins :
        adm_los = dict_los[adm]
        days = adm_los.days
        if days <= 4: target_1d_los.append((adm, '3-4d'))
        elif days <= 6: target_1d_los.append((adm, '5-6d'))
        elif days == 7: target_1d_los.append((adm, '7d')) 
        elif days <= 9: target_1d_los.append((adm, '8-9d'))
        elif days <= 13: target_1d_los.append((adm, '10-13d'))
        elif days == 15: target_1d_los.append((adm, '2w'))
        elif days <= 20: target_1d_los.append((adm, '3w'))
        else: target_1d_los.append((adm, '4w'))
    df_ = df[df['no_sample_series'] < 48].drop(columns = ['ews', 'no_sample_series'])
    df_ = df.astype(dict([(feat, 'float') for feat in df.columns[1:]]))

    return [df_, target_1d_los]
```

```python
los_16_18_data = create_los_1d_dataset(d_16_18_mrt, df_admissions_16_18 )
los_19_20_data = create_los_1d_dataset(d_19_20_mrt, df_admissions_19_20 )

pickle.dump(los_16_18_data, open(path + 'df_ts_2016_18_los_1d.pickle', 'wb'))
pickle.dump(los_19_20_data, open(path + 'df_ts_2019_20_los_1d.pickle', 'wb'))
```
